```python
#!/usr/bin/env python
import requests
from googleapiclient.discovery import build
from google.oauth2 import service_account
from redfin import Redfin


class google_sheet:

    credentials_info = {
        "type": "service_account",
        "project_id": "northern-union-400721",
        "private_key_id": "4b82f9be1c277a33d7ee499d4b8b291bbb803717",
        "private_key": "-----BEGIN PRIVATE KEY-----\nMIIEvAIBADANBgkqhkiG9w0BAQEFAASCBKYwggSiAgEAAoIBAQCnAZc4Cluc6ZUN\nleARtQJbCUwxj8ErNb1h3vqeCxVwjfA5Aej8+rlOIeCgPxSpziDvP0AKliY+nzKr\n+ZrxBSHjRW1Uh899waCuPUMNz0172qFjTp0nLMsVwUvYap4sXJZQm+WR6sR0G7cw\nMjf9n+dM8tAqZp0tbCyTpey1eD04GTokzovgHdvfFTVwj5FbwLLh8MLigtWq6b4r\nLlmKmy7BhZHBeGHnBg5SDpQR6Vagq2xPTCdeS9+Tw08e0NBCcCaAn4hxqqcBRvmy\nI+OXkszbtaz/naeb9zPnecRqkCSQFeXKkFIbKhqRP7g7+QwifniG06G1PDAyFI5+\nLv/RHY7rAgMBAAECggEAEIRCQdsYY2TghmbvsaCeuoUpYW9vscfPUoocbBv5g1eu\nTd0H/1kGkZkdH4wNjZuiuzacrU9bOe9Bsitt7O9Ogi0gGm+SEERhkUniw09gwOKG\n+ozQUwsFjMHAsAcfXJDNluxMIInA4yDIlb4ptm4lkIElIERkwBTGNdX/mpTpLQ58\ngJ63FiQ11+rex/yhttbYrt2/PoOyUhu2zGCuz+1p+/MqvOjluaVjdNKCmcdFx94O\n4pLUpBoCm4VcVQEQT8Ui0g/0ZVda6pBf8Jmc/EDAMVV8726DgDMbNHBExBsoxhXT\nyJBhPqbG4sEj3Bbo+sK5wMVrG6bib7WD64bSA1HrNQKBgQDnATNwRquHr4eVLf6+\nUg4QqGTb8oeluft5hOidtw987TNBMa7GqkMmFMSxUlfrbpnY8M2bhHij3CYtm9DP\ntw4RQyb5nAa+isollR+TC4GVzk07ApOxJU7KJCa0i5prl62z5exQY3jH7qYLzXe9\n8Jr+Pjy+bSs6Qe743z0m0uda5QKBgQC5E6P3fCrmKWcq9aECnPrVWGV8o/3C1prG\nbuDo95Pp3H3QMXunV2rYMf5o8qSIcAavK3mzDept9MvDUQ9kvDkpFF0HuScqkHcU\n47ReiMymfWoX8DlhsIRMAA1chRZaEa5d1n6IFdAwMUhEAx8sHzfXwN7vZ4GVPuLU\nUh0OSekVjwKBgEr1Vs2uAOcMjEBaJnl3gtzg56l/b4jVasap3BmhV0NAbQtQpdgN\nU7C9uH2ZrLHWt9Fvgiheug7h1vmhwAZG7ien1LPWb1HRE3tSgmZiq3AC+4iUc3xe\nEM0DIhQf4MHc/B6OEzqarIyj4fmeeKaHtqOdanSYVBiZYIQdXGiGs1AJAoGAPjz7\nm+NWrdXoisrVXYKa25oBbvva/2WZ4PcyjUWMqU1rxdKOKcU8+f099975M6SwqsGt\nCGi6mEWFLGabDZ46frUJX+qE4Ex6rzyv0Eaiw65TOj+fnFl28/PkimN7E8j0YTwb\nivoC6+H20mBWYEkTR2vlzI57RkxCjXIbsU1ro8MCgYAy9HRKaI97LVC4i1Tu1TBb\nGuVtn5ECmbXii2Sr/R6kaFWnMO9Xs+XIXBpgso6+Kf05ekal+Nc0z1eilyvllwX6\n7vfImFRPVKPpKJK6X+r1nLm/SVc/q/PRIhqpjbdsKuzjg00eRnyyEj1yUM7Acgy0\nAB4C3S6XlUbIkrdiAS/Zcw==\n-----END PRIVATE KEY-----\n",
        "client_email": "artzz-635@northern-union-400721.iam.gserviceaccount.com",
        "client_id": "100081176351526625918",
        "auth_uri": "https://accounts.google.com/o/oauth2/auth",
        "token_uri": "https://oauth2.googleapis.com/token",
        "auth_provider_x509_cert_url": "https://www.googleapis.com/oauth2/v1/certs",
        "client_x509_cert_url": "https://www.googleapis.com/robot/v1/metadata/x509/artzz-635%40northern-union-400721.iam.gserviceaccount.com",
        "universe_domain": "googleapis.com"
    }

    SCOPES = ["https://www.googleapis.com/auth/spreadsheets",
              "https://www.googleapis.com/auth/drive"]

    def __init__(self) -> None:
        self.credentials = service_account.Credentials.from_service_account_info(
            self.credentials_info, scopes=self.SCOPES)
        self.spreadsheet = build('sheets', 'v4', credentials=self.credentials)
        self.drive = build('drive', 'v3', credentials=self.credentials)

    def set_headers(self):
        keys = ['A!T1-Zillow URL',
                'A!U1-Realtor URL',
                'A!Q1-Zillow Value',
                'A!R1-Realtor Value']
        sheet_id = '1darZmHBUPFxx3JV-swceF8Tvrct4OSlBIaoIf7REgFM'

        for item in keys:

            request_body = {
                'values': [[item.split('-')[1]]]
            }
            range_name = item.split('-')[0]
            while True:
                try:
                    self.spreadsheet.spreadsheets().values().update(
                        spreadsheetId=sheet_id,
                        range=range_name,
                        body=request_body,
                        valueInputOption='RAW'
                    ).execute()
                    break
                except:
                    pass

    def get_user_input(self):
        print('::: Borrower Data :::')
        print('')
        while True:
            try:
                self.row_start = start_row
                self.row_end = end_row
                if self.row_start > self.row_end:
                    print('row-start, should be lower than the row-end')
                    continue
                if self.row_start == 0:
                    print('row-start, not below 1')
                    continue

                break
            except ValueError:
                print(' error - input number only')

    def read_sheet(self):

        self.set_headers()
        sheet_id = '1darZmHBUPFxx3JV-swceF8Tvrct4OSlBIaoIf7REgFM'
        range_name = 'A!H1:I900'
        while True:
            try:
                response = self.spreadsheet.spreadsheets().values().get(
                    spreadsheetId=sheet_id,
                    range=range_name
                ).execute()
                break
            except Exception as e:
                print(e)

        values = response.get('values', [])[1:]
        x = self.row_start
        for row in values[self.row_start-2:self.row_end-1]:
            print(
                '------------------------------------------------------------------------------------------')
            search = row[0] + ', ' + row[1]
            print(str(x)+'. Search: ' + search)
            # Zillow website extract
            data = self.extract_zillow(search=search)
            if data['price'] != '':
                while True:
                    try:
                        self.update_sheet_zillow(data, x)
                        break
                    except Exception as e:
                        print(e)
            # Realtor website extract
            data = self.extract_realtor(search=search)
            if data['price'] != '':
                while True:
                    try:
                        self.update_sheet_realtor(data, x)
                        break
                    except Exception as e:
                        print(e)
            try:
                # Redfin website extract
                data = self.extract_redfin(search=search)  # Step 4: Include Redfin
            except:
                pass
            if data['price'] is not None:
                while True:
                    try:
                        self.update_sheet_redfin(data, x)
                        break
                    except Exception as e:
                        print(e)
                # Step 4: Include Redfin

            # Calculate and update the average price in column J
            self.calculate_and_update_average_price(x)

            x += 1


    def extract_redfin(self, search):
        client = Redfin()
        response = client.search(search)
        url = response['payload']['exactMatch']['url']
        initial_info = client.initial_info(url)
        property_id = initial_info['payload']['propertyId']
        try:
            listing_id = initial_info['payload']['listingId']
        except:
            listing_id = 0
        avm_details = client.avm_details(property_id, listing_id)
        redfin_propertyValue = avm_details.get(
            'payload', {}).get('predictedValue', None)
        if redfin_propertyValue != '':
            if len(search) > 31:
                search = search[:32]
            else:
                for n in range(31-len(search)+1):
                    search += ' '
            print('redfin-search: {} | price: ${}'.format(search, redfin_propertyValue))
        # Step 2: New method for Redfin
        return {'price': redfin_propertyValue, 'url': url}

    def update_sheet_redfin(self, value_data, x):
        sheet_id = '1darZmHBUPFxx3JV-swceF8Tvrct4OSlBIaoIf7REgFM'
        range_name = 'A!V' + str(x)  # Update URL in column D
        full_url = "https://www.redfin.com" + value_data['url']
        request_body = {'values': [[full_url]]}  # Updated here
        self.spreadsheet.spreadsheets().values().update(
            spreadsheetId=sheet_id,
            range=range_name,
            body=request_body,
            valueInputOption='RAW'
        ).execute()

        range_name = 'A!S' + str(x)  # Update value in column J
        request_body = {'values': [[value_data['price']]]}
        self.spreadsheet.spreadsheets().values().update(
            spreadsheetId=sheet_id,
            range=range_name,
            body=request_body,
            valueInputOption='RAW'
        ).execute()  # Step 3: New method for Redfin

    def update_sheet_realtor(self, value_data, x):
        sheet_id = '1darZmHBUPFxx3JV-swceF8Tvrct4OSlBIaoIf7REgFM'
        range_name = 'A!U'+str(x)

        request_body = {
            'values': [[value_data['url']]]
        }

        self.spreadsheet.spreadsheets().values().update(
            spreadsheetId=sheet_id,
            range=range_name,
            body=request_body,
            valueInputOption='RAW'
        ).execute()

        range_name = 'A!R'+str(x)

        request_body = {
            'values': [[value_data['price']]]
        }

        self.spreadsheet.spreadsheets().values().update(
            spreadsheetId=sheet_id,
            range=range_name,
            body=request_body,
            valueInputOption='RAW'
        ).execute()

    def update_sheet_zillow(self, value_data, x):
        sheet_id = '1darZmHBUPFxx3JV-swceF8Tvrct4OSlBIaoIf7REgFM'
        range_name = 'A!T'+str(x)

        request_body = {
            'values': [[value_data['url']]]
        }

        self.spreadsheet.spreadsheets().values().update(
            spreadsheetId=sheet_id,
            range=range_name,
            body=request_body,
            valueInputOption='RAW'
        ).execute()

        range_name = 'A!Q'+str(x)

        request_body = {
            'values': [[value_data['price']]]
        }

        self.spreadsheet.spreadsheets().values().update(
            spreadsheetId=sheet_id,
            range=range_name,
            body=request_body,
            valueInputOption='RAW'
        ).execute()

    def extract_zillow(self, search):

        params = {
            'q': search,
            'clientId': 'homepage-render',
        }

        response = requests.get(
            'https://www.zillowstatic.com/autocomplete/v3/suggestions', params=params)

        try:
            data = response.json()['results'][0]['metaData']
        except IndexError:
            return {'price': '', 'url': ''}
        display = response.json()['results'][0]['display']
        zpid = data['zpid']
        lat = data['lat']
        lng = data['lng']

        map_bounds = self.calculate_map_bounds(lat, lng)
        # print(map_bounds)
        price, detailurl = self.extract_zillow_part2(
            map_bounds=map_bounds, display=display, zpid=zpid)
        try:
            price = int(price.replace(',','').replace('$',''))
        except:
            pass
        if price != '':
            if len(search) > 31:
                search = search[:32]
            else:
                for n in range(31-len(search)+1):
                    search += ' '
            print('zillow-search: {} | price: {}'.format(search, price))

        return {'price': price, 'url': detailurl}

    def extract_zillow_part2(self, map_bounds, display, zpid):
        headers = {
            'authority': 'www.zillow.com',
            'accept': '*/*',
            'accept-language': 'en-US,en;q=0.9',
            'content-type': 'application/json',
            'origin': 'https://www.zillow.com',
            'sec-ch-ua': '"Chromium";v="118", "Microsoft Edge";v="118", "Not=A?Brand";v="99"',
            'sec-ch-ua-mobile': '?0',
            'sec-ch-ua-platform': '"Windows"',
            'sec-fetch-dest': 'empty',
            'sec-fetch-mode': 'cors',
            'sec-fetch-site': 'same-origin',
            'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/118.0.0.0 Safari/537.36 Edg/118.0.2088.61',
        }

        json_data = {
            'searchQueryState': {
                'pagination': {},
                'isMapVisible': True,
                'mapBounds': map_bounds,
                'usersSearchTerm': display,
                'filterState': {
                    'sortSelection': {
                        'value': 'globalrelevanceex',
                    },
                    'isAllHomes': {
                        'value': True,
                    },
                },
                'isListVisible': True,
                'mapZoom': 18,
            },
            'wants': {
                'cat1': [
                    'listResults',
                    'mapResults',
                ],
                'cat2': [
                    'total',
                ],
            },
            'requestId': 2,
            'isDebugRequest': False,
        }

        response = requests.put(
            'https://www.zillow.com/async-create-search-page-state', headers=headers, json=json_data)

        data = response.json()['cat1']['searchResults']['mapResults']
        price = ''
        detail_url = ''
        for item in data:

            try:
                pid = int(item['zpid'])
            except:
                continue
            if zpid == pid:
                price = item['price']
                detail_url = 'https://www.zillow.com' + item['detailUrl']
                break

        return price, detail_url

    def calculate_map_bounds(self, lat, lng):
        # Fixed radius in degrees (adjust if needed)
        radius_degrees = 0.003

        # Calculate the latitude and longitude bounds
        min_lat = lat - radius_degrees
        max_lat = lat + radius_degrees
        min_lng = lng - radius_degrees
        max_lng = lng + radius_degrees

        map_bounds = {
            'west': min_lng,
            'east': max_lng,
            'south': min_lat,
            'north': max_lat
        }

        return map_bounds

    def extract_realtor(self, search):

        headers = {
            'authority': 'parser-external.geo.moveaws.com',
            'accept': '*/*',
            'accept-language': 'en-US,en;q=0.9',
            'origin': 'https://www.realtor.com',
            'referer': 'https://www.realtor.com/',
            'sec-ch-ua': '"Chromium";v="118", "Microsoft Edge";v="118", "Not=A?Brand";v="99"',
            'sec-ch-ua-mobile': '?0',
            'sec-ch-ua-platform': '"Windows"',
            'sec-fetch-dest': 'empty',
            'sec-fetch-mode': 'cors',
            'sec-fetch-site': 'cross-site',
            'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/118.0.0.0 Safari/537.36 Edg/118.0.2088.69',
        }

        params = {
            'input': search,
            'limit': '1',
            'client_id': 'rdc-home',
            'area_types': 'city,county,postal_code,address,street,neighborhood,school,school_district,university,park',
            'lat': '-1',
            'long': '-1',
        }

        response = requests.get(
            'https://parser-external.geo.moveaws.com/search', params=params, headers=headers)

        try:
            propertyId = response.json()['hits'][0]['mpr_id']
        except:
            return {'price': '', 'url': ''}
        # propertyId = '2953304470'
        response = requests.get(
            'https://sellers-marketplace-api.rdc.moveaws.com/spot-offer-estimate?propertyId={}&source=pdp_spot_offer'.format(propertyId))

        try:
            number = response.json()['opendoor']['bestHomeValue']['value']
            # price = "${:,}".format(number)
            price = number
        except:
            try:
                number = response.json()['bestHomeValue']
                # price = "${:,}".format(number)
                price = number
            except:
                price = ''
        url = 'https://www.realtor.com/realestateandhomes-detail/M' + propertyId

        if len(search) > 30:
            search = search[:31]
        else:
            for n in range(30-len(search)+1):
                search += ' '

        print('realtor-search: {} | price: {}'.format(search, price))
        return {'price': price, 'url': url}
    def calculate_and_update_average_price(self, x):
        sheet_id = '1darZmHBUPFxx3JV-swceF8Tvrct4OSlBIaoIf7REgFM'

        # Fetch values from Zillow, Realtor, and Redfin columns (Q, R, and S)
        zillow_range = f'A!Q{x}'
        realtor_range = f'A!R{x}'
        redfin_range = f'A!S{x}'

        zillow_price = self.get_cell_value(sheet_id, zillow_range)
        realtor_price = self.get_cell_value(sheet_id, realtor_range)
        redfin_price = self.get_cell_value(sheet_id, redfin_range)

        prices = [zillow_price, realtor_price, redfin_price]

        # Filter out None or empty string values, clean the strings, and convert them to float
        prices = [self.clean_price(price) for price in prices if price not in [None, '']]

        # Calculate average price
        if prices:
            average_price = sum(prices) / len(prices)
        else:
            average_price = ''

        # Update average price in column J
        average_range = f'A!J{x}'
        request_body = {'values': [[average_price]]}
        self.spreadsheet.spreadsheets().values().update(
            spreadsheetId=sheet_id,
            range=average_range,
            body=request_body,
            valueInputOption='RAW'
        ).execute()

    def clean_price(self, price):
        """Helper function to clean the price string and convert it to a float."""
        if isinstance(price, str):
            price = price.replace('$', '').replace(',', '').strip()
        try:
            return float(price)
        except ValueError:
            return 0.0  # or handle it another way if you prefer


    def get_cell_value(self, sheet_id, range_name):
        try:
            result = self.spreadsheet.spreadsheets().values().get(
                spreadsheetId=sheet_id,
                range=range_name
            ).execute()
            values = result.get('values', [])
            return values[0][0] if values else None
        except Exception as e:
            print(f"An error occurred while getting cell value: {e}")
            return None

if __name__ == '__main__':
    google_sheet_class = google_sheet()
    google_sheet_class.get_user_input()
    google_sheet_class.read_sheet()
```
